```python
import numpy as np
import pandas as pd
import seaborn as sb
from numpy import sqrt, abs, round
import scipy.stats as stats
from scipy.stats import norm
```

```python

def twoSampZ(X1, X2, mudiff, sd1, sd2, n1, n2):
    pooledSE = sqrt(sd1**2/n1 + sd2**2/n2)
    z = ((X1 - X2) - mudiff)/pooledSE
    pval = 2*(1 - norm.cdf(abs(z)))
    return round(z,3), pval
```

```python
data
Malemean=data.loc[data['sex']=='Male','salary'].mean()
Femalemean=data.loc[data['sex']=='Female','salary'].mean()
Malesd=data.loc[data['sex']=='Male','salary'].std()
Femalesd=data.loc[data['sex']=='Female','salary'].std()
Malenum=data.loc[data['sex']=='Male','salary'].count()
Femalenum=data.loc[data['sex']=='Female','salary'].count()

z,p=twoSampZ(Malemean, Femalemean, 0, Malesd, Femalesd, Malenum, Femalenum)
print ('Z =',z,'p =',p)
```

```python
print (Malemean, Femalemean)
```

```python
if p<0.05:
    print("we reject null hypothesis")
else:
    print("we accept null hypothesis")
```
