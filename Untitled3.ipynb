```python
x=[1,2,3,4,5, 6, 7,-5, -30,25, 0, 40,100]
s=0
for i in range (len(x)):
    if x[i]<0:
        j=i
        break
        
for k in range (1, j):
    
    if x[k-1]<x[k]:
        s+=1
        
    else:
        break
    


if s==j-1:
    print ("increasing  series")
else:
    print("not increasing series")
        
```
