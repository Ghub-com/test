```python
import numpy as np
import random as rn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
r=[]
a=[]
b=[]
c=[]

rn.seed()
def search(n,xl,xu,yl,yu):

    
   
    
    for i in range(0,n):
        
        r.append(rn.uniform(-1,1))
    
        x=xl+(xu-xl)*r[i]
        a.append(x)
        y=yl+(yu-yl)*r[i]
        b.append(y)
       
        c.append(2*(a[i])*(b[i])+2*(a[i])-(b[i])**2-2*(b[i])**2)
        max1=c[0]
        max2=a[0]
        max3=b[0]
        if (max1<c[i]):
            max1=c[i]
            max2=a[i]
            max3=b[i]
    return max1
```

```python
def error(n):
    er=[]
    it=[]
    for z in range (0,n):
        it.append(z)
        
        error=abs((2-search(z+1,0,3,0,2))/2) 
        er.append(error)
    return it,er
```

```python
it2,er2=error(100)
plt.plot(it2,er2)
plt.yscale('log')
plt.title('Random search error propagation')
plt.ylabel("error")
plt.xlabel("iteration number")
plt.show()
```
