```python
n = int(input())
 
factorial = 1
while n > 1:
    factorial *= n
    n -= 1
 
print(factorial)
```

```python
n = int(input())
 
factorial = 1
 
for i in range(1, n+1):
    factorial *= i
 
print(factorial)
```

```python
fib1 = 1
fib2 = 1
 
n = input("Номер элемента ряда Фибоначчи: ")
n = int(n)
 
i = 0
while i < n - 2:
    fib_sum = fib1 + fib2
    fib1 = fib2
    fib2 = fib_sum
    i = i + 1
 
print("Значение этого элемента:", fib2)
```

```python
a = int(input())
b = int(input())
 
while a != 0 and b != 0:
    if a > b:
        a = a % b
    else:
        b = b % a
 
print(a + b)
```

```python
def gcd(a, b):
    while a != b:
        if a > b:
            a = a - b
        else:
            b = b - a        
    print(a)
gcd(20, 220)    
```

```python
print ("the number to check:\n")
n=int(input())
i=1
count=0
while i < n+1:
    if n % i == 0:
        count+=1
    i+=1
if count>2:
    print ("not prime")
else:
    print ("prime")
```
